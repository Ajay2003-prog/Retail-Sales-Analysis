```python
import pandas as pd
```

# UPLOADING Files


```python
brands = pd.read_csv('brands.csv')
categories = pd.read_csv('categories.csv')
customers = pd.read_csv('customers.csv')
orders = pd.read_csv('orders.csv')
order_items = pd.read_csv('order_items.csv')
products = pd.read_csv('products.csv')
staffs = pd.read_csv('staffs.csv')
stocks = pd.read_csv('stocks.csv')
stores = pd.read_csv('stores.csv')

```

# CHECKING NULLS


```python
print(brands.isnull().sum())
print(categories.isnull().sum())
print(customers.isnull().sum())
print(orders.isnull().sum())
print(order_items.isnull().sum())
print(products.isnull().sum())
print(staffs.isnull().sum())
print(stores.isnull().sum())
print(stocks.isnull().sum())
```

# CHecking Duplicates

```python
print("Brands duplicates:", brands.duplicated().sum())
print("Categories duplicates:",categories.duplicated().sum())
print("Customer duplicates:",customers.duplicated().sum())
print("Order_items duplicates:",order_items.duplicated().sum())
print("Products duplicates:",products.duplicated().sum())
print("Staffs duplicates:",staffs.duplicated().sum())
print("Orders duplicates:",orders.duplicated().sum())
print("Stores duplicates:",stores.duplicated().sum())
print("Stocks duplicates:",stocks.duplicated().sum())
```

# Droping Duplicates



```python
brands.drop_duplicates(inplace=True)
categories.drop_duplicates(inplace=True)
customers.drop_duplicates(inplace=True)
order_items.drop_duplicates(inplace=True)
products.drop_duplicates(inplace=True)
staffs.drop_duplicates(inplace=True)
orders.drop_duplicates(inplace=True)
stores.drop_duplicates(inplace=True)
stocks.drop_duplicates(inplace=True)
```

# Removing Extra Spaces

```python
brands = brands.apply(lambda x: x.str.strip() if x.dtype == "object" else x)
categories = categories.apply(lambda x: x.str.strip() if x.dtype == "object" else x)
customers = customers.apply(lambda x: x.str.strip() if x.dtype == "object" else x)
order_items = order_items.apply(lambda x: x.str.strip() if x.dtype == "object" else x)
products = products.apply(lambda x: x.str.strip() if x.dtype == "object" else x)
staffs = staffs.apply(lambda x: x.str.strip() if x.dtype == "object" else x)
orders = orders.apply(lambda x: x.str.strip() if x.dtype == "object" else x)
stores = stores.apply(lambda x: x.str.strip() if x.dtype == "object" else x)
stocks = stocks.apply(lambda x: x.str.strip() if x.dtype == "object" else x)
```

# Standradize Text

```python
brands = brands.apply(lambda x: x.str.title() if x.dtype == "object" else x )
categories = categories.apply(lambda x: x.str.title() if x.dtype == "object" else x )
customers = customers.apply(lambda x: x.str.title() if x.dtype == "object" else x )
order_items = order_items.apply(lambda x: x.str.title() if x.dtype == "object" else x )
products = products.apply(lambda x: x.str.title() if x.dtype == "object" else x )
staffs = staffs.apply(lambda x: x.str.title() if x.dtype == "object" else x )
orders = orders.apply(lambda x: x.str.title() if x.dtype == "object" else x )
stores = stores.apply(lambda x: x.str.title() if x.dtype == "object" else x )
stocks = stocks.apply(lambda x: x.str.title() if x.dtype == "object" else x )
```

# Discount

```python
if 'discount' in order_items.columns:
  order_items['discount'].fillna(0, inplace=True)
```

# Checking data types

```python
print(brands.dtypes)
print(categories.dtypes)
print(customers.dtypes)
print(order_items.dtypes)
print(products.dtypes)
print(staffs.dtypes)
print(orders.dtypes)
print(stores.dtypes)
print(stocks.dtypes)
```

# Date Format

```python
orders['order_date'] = pd.to_datetime(orders['order_date'])
orders['shipped_date'] = pd.to_datetime(orders['shipped_date'])
orders['required_date'] = pd.to_datetime(orders['required_date'])
orders.info()
```

# Saving Files

```python
brands.to_csv('brands_cleaned.csv', index=False)
categories.to_csv('categories_cleaned.csv', index=False)
customers.to_csv('customers_cleaned.csv',index=False)
order_items.to_csv('order_items_cleaned.csv',index=False)
products.to_csv('products_cleaned.csv',index=False)
staffs.to_csv('staffs_cleaned.csv',index=False)
orders.to_csv('orders_cleaned.csv',index=False)
stores.to_csv('stores_cleaned.csv',index=False)
products.to_csv('products_cleaned.csv',index=False)
stocks.to_csv('stocks_cleaned.csv',index=False)
```
